```python
def train_main():
    win_1 = []
    win_2 = []
    noise = AdaptiveParamNoiseSpec(initial_stddev=0.5)
    dqn_1 = dqn_agent(
        target_update=20, input_dim=3, lam=0.6, gamma=.9,
        lr=1e-4, noise=noise, eps_decay=500, eps_start=.7)
    loss = dqn_1.init_param_optim(150)
    #param = torch.load('dqn_param2.pt')
    #dqn_1.policy_model.load_state_dict(param)
    #dqn_2 = dqn_agent(input_dim=3, lam=0.6, gamma=.9, lr=1e-4)
    
    random_agent = randomAgent()
    greedy_agent = greedyAgent()
    memory = Replay_Memory(400)

    for i in tqdm(range(1000)):
        if i%4==0:
            dqn_2 = copy.deepcopy(dqn_1)
        tmp_w_1 = 0
        tmp_w_2 = 0
        agent_2 = np.random.choice([dqn_2, greedy_agent, random_agent], p=[0.5,0.25,0.25])
        
        data_1, data_2 = get_play_data(dqn_1 , agent_2)
        memory.push(data_1)
        memory.push(data_2)
        tmp_w_1+=float(data_1['rewards'][0])
        tmp_w_2+=float(data_2['rewards'][0])

        win_1.append(tmp_w_1)    
        win_2.append(tmp_w_2)    
        dqn_1.optimize_model(memory)
        
    win_1_df = pd.DataFrame()
    win_1_df['last_reward'] = win_1
    win_1_df['last_reward'] = win_1_df['last_reward'].rolling(10,min_periods=1).mean()
    win_1_df['result'] = (win_1_df['last_reward']>0).astype(int)
    win_1_df['result'] = win_1_df['result'].rolling(10,min_periods=1).mean()
    
    loss_df = pd.DataFrame()
    loss_df['dqn_loss'] = dqn_1.loss_1_list
    loss_df['dqn_loss'] = loss_df['dqn_loss'].rolling(10,min_periods=1).mean()
    
    return dqn_1, win_1_df, loss_df
```

```python
dqn_1, win_1_df, loss_df = train_main()
```
